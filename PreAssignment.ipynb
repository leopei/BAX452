```python
X = df[['PRCP','TMIN']]
Y = df['TMAX']
```

```python
from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size= 0.2, random_state=1007)
```

```python
from sklearn.linear_model import LinearRegression
lrm = LinearRegression()
```

```python
lrm.fit(X_Train, Y_Train)
```

```python
weights = pd.DataFrame(lrm.coef_, X.columns, columns=['Weights'])
weights
```

```python
projections = lrm.predict(X_Test)
```

```python
plt.scatter(Y_Test,projections)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(Y_Test, projections))
print('MSE:', metrics.mean_squared_error(Y_Test, projections))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_Test, projections)))
```

```python
predict_train = lrm.predict(X_Train)
predict_test = lrm.predict(X_Test)

print('Training Accuracy:', metrics.accuracy_score(Y_Train, projections))
print('Test Accuracy:', metrics.accuracy_score(Y_Test, projections))
```
